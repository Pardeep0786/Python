```python
#Gives the labels to pie chart
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Litchi"]
plt.pie(y, labels = myLabels)
plt.show()
```

```python
#Start the first wedge at 90 degrees
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Litchi"]
plt.pie(y, labels = myLabels, startangle = 90)
plt.show()
```

```python
#Pull the "Watermelon" wedge 0.2 from the center of the pie
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Watermelon"]
myexplode = [0.0, 0.0, 0.0, 0.2]

plt.pie(y, labels = myLabels, startangle = 90, explode = myexplode)
plt.show()
```

```python
#Provide the shadow with the help of "shadow" keyword
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Watermelon"]
myexplode = [0.0, 0.0, 0.0, 0.2]

plt.pie(y, labels = myLabels, startangle = 90, explode = myexplode, shadow = True)
plt.show()
```

```python
#Provide the colors to the chart
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Watermelon"]
myexplode = [0.0, 0.0, 0.0, 0.2]
mycolors = ["c", "m", "hotpink", "red"]

plt.pie(y, labels = myLabels, startangle = 90, explode = myexplode, shadow = True, colors = mycolors)
plt.show()
```

```python
#To add a list of explanation for each wedge
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Watermelon"]
myexplode = [0.0, 0.0, 0.0, 0.2]
mycolors = ["c", "m", "hotpink", "red"]

plt.pie(y, labels = myLabels, startangle = 90, explode = myexplode, shadow = True, colors = mycolors)
plt.legend()
plt.show()
```

```python
#To change the position of the legend
y = np.array([10, 55, 25, 10])
myLabels = ["Apple", "Banana", "Cherry", "Watermelon"]
myexplode = [0.0, 0.0, 0.0, 0.2]
mycolors = ["c", "m", "hotpink", "red"]

plt.pie(y, labels = myLabels, startangle = 90, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title = "Four fruits", loc = 'center left')
plt.show()
```
