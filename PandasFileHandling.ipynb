```python
#import pd library
import pandas as pd 
import warnings                        #import warnings library for avoid warnings
warnings.filterwarnings('ignore')   
import matplotlib.pyplot as plt         #import pyplot of matplotlib for making graphs
```

```python
#Load a csv file
df = pd.read_csv("NEW.csv")
```

```python
#To check the first five records of the dataset 
df.head()
```

```python
#To check the last five records of the dataset
df.tail()
```

```python
#To check the null values in the dataframe
df.isnull().sum()
```

```python
#To fill the null values
df['Class'].fillna('BCA', inplace = True)
```

```python
df.head()
```

```python
df.tail()
```

```python
#Fill the null value with mode 
x = df['City'].mode()[0]
df['City'].fillna(x, inplace = True)
```

```python
df.head()
```

```python
#To access the particular row 
df.loc[1, 'Age'] = 18
```

```python
df.head()
```

```python
#To check the duplicates
df.duplicated()
```

```python
#For removing the duplicates 
df.drop_duplicates(inplace = True)
```

```python
#Plot diagram of Roll no.
df.plot()
plt.show()
```

```python
#Scatter plot
df.plot(kind = 'scatter', x = 'Roll no.', y = 'Salary')
plt.show()
```
